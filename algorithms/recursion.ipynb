#### **Problem-1 :**
Write a recursive function to count the number of items in a list

```python
def count_items(arr):
  """count of elements in array
  Recursively counts number of elements in an array
  Time Complexity: O(n^2)
  """
  # what to argument every recursive call
  # what do it return ?

  new_array = list(arr) # O(n) create shallow copy so original array is not affected
  count = 0
  if len(new_array) > 0: # if len(arr) > 0
    new_array.pop(0) # O(n)
    # print("popped!!.")
    return 1 + count_items(new_array)
  else:
    return 0

array = []
result = count_items(array)
result, array
```

```python
# The below function does not modify the original array

def count_items_2(array):
  """count of elements in array
  Recursively counts number of elements in an array
  Time Complexity: O(n^2)
  """
  # what to argument every recursive call
  # what do it return ?

  if len(array) > 0:
    return 1 + count_items_2(array[1:])
  else:
    return 0

array = [1,2,5,4,6]
result = count_items_2(array)
result, array
```

#### **Problem-2 :**
Write a recursive function to find the maximum number in a
list.

```python
def get_max_from_list(input_array, index= 0):
  """maximum of an array
  Recursively finds the maximum number in an array
  Time Complexity: O(n^2)
  """
  # what do I feed to every recursive call ?.
  # what do I return from every recursive call ?.

  array = list(input_array) # O(n)
  if len(array) > 1:
    if array[index] < array [index + 1]:
      array.pop(index) # O(n)
      return get_max_from_list(array, index)
    elif array[index] > array [index + 1]:
      array.pop(index + 1) # O(n)
      return get_max_from_list(array, index)
  else:
    return array[0]

imp_array = [1,2,88,5,4,6,55,0,99,81]
result = get_max_from_list(imp_array)
result, array
```

```python
def get_max_from_list_2(input_array, index= 0):
  """maximum of an array
  Recursively finds the maximum number in an array
  Time Complexity: O(n^2)
  """

  if input_array[index] < input_array [index + 1]:
    return get_max_from_list(input_array[1:], index)
  elif input_array[index] > input_array [index + 1]:
    index_elem = input_array[index]
    input_array[index] = input_array [index + 1]
    input_array[index + 1] = input_array [index_elem]
    return get_max_from_list(input_array[1:], index)
  else:
    return array[0]

imp_array = [1,100,2,88,5,4,6,55,0,99,81]
result = get_max_from_list(imp_array)
result, imp_array
```

```python
def get_max_from_list_3(array):
    """Recursively finds the maximum number in an array using pairwise comparison."""
    if len(array) == 1:
        return array[0]  # Base case: only one item left

    if array[0] < array[1]:
        return get_max_from_list_3(array[1:])  # Drop the smaller one
    else:
        return get_max_from_list_3([array[0]] + array[2:])  # Keep the larger one

imp_array = [1,100,2,88,5,4,6,55,0,99,81]
result = get_max_from_list_3(imp_array)
result, imp_array
```
