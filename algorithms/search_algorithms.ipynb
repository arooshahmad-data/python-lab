## Linear Search Algorithm

```python
data = [1,2,4,3,6,7,8,9,0,5]

def linear_search(arr, target):
  for entry in arr:
    if entry == target:
      return True
  return False

linear_search(data,10)
```

## Binary Search

```python
import math
def binary_search(arr, target):
  """
  get length and divide by 2 get element at index if found return True else return False
  if not found create new array upto the index then again seach the center by taking length and dividing by 2
  """

  if arr == []:
    return False

  arr = sorted(arr) # sort array
  print("Array:",arr)
  c_arr_len  = len(arr) # current array length
  i_to_search = int(math.floor(len(arr)) / 2) # index to search for in array
  print(f"Index : {i_to_search} - Array {arr[i_to_search]}")

  if arr[i_to_search] == target:
    return True

  elif arr[i_to_search] > target:
    new_arr = arr[:i_to_search]
    return binary_search(new_arr, target)

  elif arr[i_to_search] < target:
    new_arr = arr[i_to_search + 1:]
    return binary_search(new_arr, target)

  return False

# data = [1,2,3,4,5,6,7,8,9,0]

binary_search(data, 15)
```

The above implementation is recursive which has space complexity as O(log n) <br>
Below is iterative implementation of **Binary Search** with space complexity O(1) <br>

Number of steps will be : log n = x,
* x will be the number of steps needed <br>
and the log is with base 2 as binary search divides the array in 2 parts everytime.

```python
def binary_search(array, target):
  """
    Iteratively perform binary search on an array
    return index of target if found else return None
  """
  low = 0
  high = len(array)-1

  while low <= high: # Edge Case: if high and low are same index then (low + high) // 2 will be same index
    mid = (low + high) // 2 # divide and floor the answer e.g. 5 // 2 = 2.5 , floor(2.5) = 2.
    guess =  array[mid]
    if guess == target:
      return mid
    elif guess > target:
      high = mid - 1
    else:
      low = mid + 1

  return None

# 1,2,3,4,5
```

```python
data = [1,2,3,4,5,6,7,8,9,0, 11]

index = binary_search(data, 11)
print(f"Fount target at index {index} in array.")
```

## Breadth First Search

Time Complexity: O (V + E), V(vertices), E(Edges)

It is a graph based algorithm. Graph can be made using hash tables to store neighbours (dictionaries in python).

```python
from collections import deque

def bfs(graph, root):
  """
  Breadth First Search
  root: root node given
  """
  search_queue = deque([root])
  # to keep track of visited nodes
  searched = set() # do not preserve order of element addition
  # use list as set do not preserve the order in which items are added to it
  searched = list()

  while search_queue:
    vertex = search_queue.popleft()
    if not vertex in searched:
      # Add only *unvisited* neighbors to the queue
      for neighbor in graph[vertex]:
          if neighbor not in searched:
             search_queue.append(neighbor)
      searched.append(vertex)

  return searched
```

```python
graph = dict()
graph['1'] = {'2','6'}
graph['2'] = {'1', '3', '4'}
graph['3'] = {'2'}
graph['4'] = {'2', '5'}
graph['5'] = {'4','8'}
graph['6'] = {'7','9'}
graph['7'] = {'6', '8'}
graph['8'] = {'5', '7'}
graph['9'] = {'6'}
```

```python
traversal_order = bfs(graph, '1')
print(traversal_order)
```
